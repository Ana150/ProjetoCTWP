```python
import speech_recognition as sr
import pyttsx3
import csv
import webbrowser
import time
import os

recognizer = sr.Recognizer()
engine = pyttsx3.init()

def obter_horario():
    hora_atual = time.strftime("%H:%M")
    return hora_atual

# Função para falar uma mensagem
def falar(mensagem):
    engine.say(mensagem)
    engine.runAndWait()
# Função para listar compromissos do dia
def listar_compromissos(data):
    compromissos = []
    if os.path.exists('agenda.csv'):
        with open('agenda.csv', mode='r') as file:
            reader = csv.reader(file)
            for row in reader:
                if len(row) >= 2 and row[0] == data:
                    compromissos.append(row[1])
    return compromissos

# Mensagem de boas-vindas (exibida apenas uma vez)
falar("Olá! Como posso ajudar você hoje? Estou aqui para fornecer informações sobre o horário atual, ajudar você a agendar compromissos, listar seus compromissos do dia ou até mesmo mostrar a previsão do tempo. Por favor, diga o que você deseja fazer.")

while True:
    with sr.Microphone() as source:
        print("Diga um comando. Diga sair para encerrar o assistente!")
        audio = recognizer.listen(source)
        if "sair" in comando:
            falar("Encerrando o assistente de voz.")
            break
    try:
        comando = recognizer.recognize_google(audio, language="pt-BR")
        print("Você disse: " + comando)

        if "horário" in comando:
            horario_atual = obter_horario()
            falar("Aqui está o horário atual. Agora são " + horario_atual)
        elif "adicionar compromisso" in comando:
            falar("Por favor, diga a data (formato aaaa-mm-dd) e o compromisso.")
            with sr.Microphone() as source:
                audio = recognizer.listen(source)
            entrada = recognizer.recognize_google(audio, language="pt-BR")
            data, compromisso = entrada.split(" ", 1)
            adicionar_compromisso(data, compromisso)
            falar("Compromisso adicionado com sucesso.")
        elif "compromissos do dia" in comando:
            data_hoje = time.strftime("%Y-%m-%d")
            compromissos = listar_compromissos(data_hoje)
            if compromissos:
                falar("Compromissos para hoje:")
                for compromisso in compromissos:
                    falar(compromisso)
            else:
                falar("Você não tem compromissos para hoje.")
        elif "previsão do tempo" in comando:
            webbrowser.open("https://www.google.com/search?q=previs%C3%A3o+do+tempo")
        elif "sair" in comando:
            falar("Encerrando o assistente de voz.")
            break
        else:
            falar("Comando não reconhecido. Tente novamente.")

    except sr.UnknownValueError:
        falar("Não entendi o comando.")
    except sr.RequestError as e:
        falar("Erro na requisição: {0}".format(e))
```
